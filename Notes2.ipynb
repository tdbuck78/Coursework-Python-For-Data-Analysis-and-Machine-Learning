```python
import pandas as pd
```

```python
df = pd.read_html('x2_actions.html')  #reading in file from HTML table
```

```python
df = df[0]   #setting first table from the table array as dataframe
```

```python
df.columns = df.iloc[0]   #fixing column formatting
```

```python
df.drop(df.index[0], inplace=True)   #removing duplicate row
```

```python
df.reset_index(drop=True)   #resetting indices
```

```python
lst = df[df['Hours'].isnull()].index.values.tolist()   # removing any null-hour rows
lst.sort(reverse = True)
for row in lst:
    df.drop(df.index[row]-1, inplace=True)  
```

```python
df['Hours'] = df['Hours'].apply(pd.to_numeric)   #making hours type int
```

```python
df['Hours'] = df['Hours']/3600
```

```python
df.to_excel('x2_actions.xlsx',sheet_name='Sheet1')
```
