```python
import datetime
import time 
import numpy as np

# Getting an array of months in unix time
unix_months = []
for year in range (2017,2018):
    for month in range(1,13):
        dt = datetime.datetime(year, month, 1)
        unix_months.append(time.mktime(dt.timetuple()))

# Don't forget january of 2018
dt = datetime.datetime(2018, 1, 1)
unix_months.append(time.mktime(dt.timetuple()))

# Making x-axis labels
x_months_arr = []
for year in range (2017,2018):
    for month in range(1,13):
        dt = datetime.datetime(year,month,1)
        x_months_arr.append(dt)
x_months = np.array(x_months_arr)


fig = plt.figure()
axes = fig.add_axes([1,0.1,10,5])

matplotlib.rcParams.update({'font.size': 100})

for provider in ['Kamau','wgarrick','Claire','trevor']:
    temp = []
    for i in range (0,len(unix_months)-1):
        temp.append( df[
        (df['Provider']== provider) & (df['Start']>unix_months[i])  & (df['Start']<unix_months[i+1])
                            ] ['Hours'].sum()
                         )
    y = np.array(temp)
    axes.plot(x_months,y,label=provider)

ax.legend()
```
